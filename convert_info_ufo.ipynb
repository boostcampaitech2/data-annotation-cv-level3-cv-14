```python
# ufo format으로 변경한 download data를 저장할 새로운 dict 생성 
new_json = {}
new_json['images'] = {}
print(new_json)
```

```python
def make_points_list(points):
    '''
    input: [x1,y1,x2,y2,...xn,yn]
    output: [[xn,yn],...] 이중 리스트 형태
    비고: string 형태로 표현된 좌표를 float로 변환
    '''
    point_cnt = len(points)//2
    if point_cnt%2 !=0: 
        return None #홀수개 point 불허
    points_list = []
    
    for i in range(point_cnt):
        x = float(points[i*2])
        y = float(points[i*2+1])
        points_list.append([x,y])
    
    return points_list
    
make_points_list(['512.20', '137.00', '547.92', '139.60', '544.85', '150.36', 
                  '536.91', '152.67', '521.12', '152.41', '517.79', '163.51'])
```

```python
img_len = len(json_file['images']) #data length=3000

for img_idx in range(img_len):
    img_info = json_file['images'][img_idx]
    for key,value in img_info.items():
        img_removed = 0
        #image level info
        if key=='name':
            image_name = value
            new_json['images'][image_name] = {}
            new_json['images'][image_name]['paragraphs'] = {}
            new_json['images'][image_name]['words'] = {}
            new_json['images'][image_name]['chars'] = {}
            new_json['images'][image_name]['tags'] = {}
            new_json['images'][image_name]['relations'] = {}
            new_json['images'][image_name]['license_tag'] = {}
            
        if key=='objects':
            word_num = len(img_info[key]) #word 개수
            #word level info
            for idx in range(word_num):
                if img_removed==1: break
                new_json['images'][image_name]['words'][str(idx)]={}
                for key2,value2 in img_info[key][idx].items():
                    if key2=='points':
                        points_list = make_points_list(value2)
                        if points_list!=None:
                            new_json['images'][image_name]['words'][str(idx)]['points'] = points_list
                        else:
                            del new_json['images'][image_name]
                            img_removed = 1
                            break
                    if key2=='Reading Direction': 
                        new_json['images'][image_name]['words'][str(idx)]['orientation'] = value2
                    if key2=='Description': 
                        new_json['images'][image_name]['words'][str(idx)]['transcription'] = value2
                        #ko,en,others 판단 후 new_json['images'][image_name]['words'][str(idx)]['language']에 넣기
                        new_json['images'][image_name]['words'][str(idx)]['language'] = ["ko"] #임시
                    if key2=='occlusion':
                        new_json['images'][image_name]['words'][str(idx)]['word_tags']='occlusion' if value2=='true' else None
                    if key2=='Don\'t Care': 
                        new_json['images'][image_name]['words'][str(idx)]['illegibility']=True if value2=='true' else False
                    #if key2=='Special Characters' and value2=='true': 
                        #특수문자 여부
                    #if key2=='Art Characters' and value2=='true': 
                        #타이포그래피 여부

#save converted result                        
with open(save_path, 'w') as fp:
    json.dump(new_json, fp)
    
```
